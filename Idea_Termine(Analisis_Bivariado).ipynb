```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Scatterplot
fig, ax = plt.subplots()
ax.scatter(df['tempo'], df['danceability'], alpha=0.4)
ax.set_title('Tempo vs. Danceability')
ax.set_xlabel('Tempo')
ax.set_ylabel('Danceability')
```

```python
#Dar un count plot
#Muestra sencilla de canciones con danceabilidad mayor al 0.5
sns.countplot(df["danceability"]>0.5,hue=df["tempo"]>122)
plt.show()
```

```python
#Scatterplot
fig, ax = plt.subplots()
ax.scatter(df['energy'], df['danceability'], alpha=0.3)
ax.set_title('Energy vs. Danceability')
ax.set_xlabel('Energy')
ax.set_ylabel('Danceability')
```
